```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('F:\GIT\python-MachineLearningInAction\PLA\data\data2.csv',header=None)
x=data.iloc[:,:2].values
y = data.iloc[:,2].values
#print(data)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:50,0],x[:50,1],color='blue',marker='*',label='Positive')
plt.scatter(x[50:,0],x[50:,1],color='red', marker='x', label='Negative')
plt.xlabel('feature1')
plt.ylabel('feature2')
plt.legend(loc = 'upper left')
plt.title('Original Data')
plt.show()
```

```python
#归一化    axis=0 按列求均值，得到两个特征的均值   axis=1 按行求均值
u=np.mean(x,axis=0) 
#print(len(u))
v=np.std(x,axis=0)
#print(v)
x=(x-u)/v
#print(x)
```

```python
plt.scatter(x[:50,0],x[:50,1],marker='*',color='blue',label='Positive')
plt.scatter(x[50:,0],x[50:,1],marker='+',color='red',label='Negitive')
plt.legend(loc = 'upper left')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.title('data2')
plt.show()
```

```python
X=np.hstack((np.ones([x.shape[0],1]),x))
w=np.random.randn(3,1)
#print(x)
#print(w)
x1=-2
y1=-(w[0]*1+w[1]*x1)/w[2]
x2=2
y2=-(w[0]*1+w[1]*x2)/w[2]
plt.scatter(x[:50,0],x[:50,1],marker='*',color='blue',label='Positive')
plt.scatter(x[50:,0],x[50:,1],marker='+',color='red',label='Negitive')
plt.plot([x1,x2],[y1,y2])
plt.legend(loc = 'upper left')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.title('data2')
plt.show()
```

```python
w = np.random.randn(3,1)
s=np.dot(X,w)
Y=y.reshape(1,100)
p=np.multiply(s,y.reshape(100,1))
t=np.where(p<0)[0]
nums=len(t)
print('初始分类错误的点个数：%2d' %(nums))
for i in range(100): 
    r = np.random.choice(nums)
    w2=y[t[r]] * X[t[r], :].reshape((3,1))
    s=np.dot(X,w2)
    p=np.multiply(s,y.reshape(100,1))
    t=np.where(p<0)[0]
    nums2=len(t)
    if nums==0:
        print('w=',w)
        break
    elif nums2<=nums:
        w = w2
        nums=nums2
        print('w=',w,)
        print('nums2=',nums2)
    else:
        print('++++++++++++++++++++++++++++++++++++++++++++++++')
        continue
print(w)
```
